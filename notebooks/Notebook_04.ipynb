### Studying the network solution at two time instants

Here we compare the network solution of the diffusion equation to the exact one. 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os, sys
sys.path.append(os.path.join(os.getcwd(), "../"))
import plotter
```

```python
predictions = np.load("../1M_Model_output.npy")

ts = np.linspace(0,1,100)
x = np.linspace(0,1,100)
```

```python

fig1 = plotter.plot_fd_solutions(predictions, x, ts, t1_index, t2_index)
cwd = os.getcwd()
fig1.savefig(cwd + '/../Plots/Model_1M.pdf')

plt.show()
```
